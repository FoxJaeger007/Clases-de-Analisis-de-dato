```python
# El primer ejercicio de práctica deberás realizar el código usando las funciones y bibliotecas
#Matplotlib y Seaborn en Python, con el respectivo grafico según el caso, escribir el respectivo
#código y mostrar el resultado, aplicar los temas vistos en clase.



import numpy as np
import matplotlib.pyplot as plt

# Definir las ciudades y los productos
ciudades = ['Ciudad A', 'Ciudad B', 'Ciudad C','Cali','Bucaramanga' ]
productos = ['Producto A', 'Producto B', 'Producto C']

# Datos de ventas específicos para cada producto en cada ciudad
ventas_producto_a = [45, 30, 50,25,25]  # Ventas del Producto A en Ciudad A, Ciudad B, Ciudad C,'Cali','Bucaramanga'
ventas_producto_b = [60, 25, 70,30,20]  # Ventas del Producto B en Ciudad A, Ciudad B, Ciudad C,'Cali','Bucaramanga'
ventas_producto_c = [25, 40, 45,20,40]  # Ventas del Producto C en Ciudad A, Ciudad B, Ciudad C,'Cali','Bucaramanga'

# Configurar el gráfico de barras
fig, ax = plt.subplots(figsize=(10, 6))
bar_width = 0.2
index = np.arange(len(ciudades))

# Crear barras para cada producto
plt.bar(index, ventas_producto_a, bar_width, label='Producto A', edgecolor='black')
plt.bar(index + bar_width, ventas_producto_b, bar_width, label='Producto B', edgecolor='black')
plt.bar(index + 2 * bar_width, ventas_producto_c, bar_width, label='Producto C', edgecolor='black')

# Añadir títulos y etiquetas
ax.set_xlabel('Ciudad')
ax.set_ylabel('Ventas')
ax.set_title('Ventas por Ciudad y Producto')
ax.set_xticks(index + bar_width)
ax.set_xticklabels(ciudades)
plt.legend(title='Producto')

# Mostrar el gráfico
plt.show()





```

```python
#Practica 2 ejercicio 2 -	Realizar un gráfico personalizado para aplicar un estilo atractivo,
# se requiere visualizar las ventas mensuales de tres tipos de vehículos (SUV, Sedan, y Hatchback)
# durante un año.



import numpy as np
import matplotlib.pyplot as plt

# Definir las ciudades y los productos
Meses =['Enero','Febrero','Marzo','Abril','Mayo','Junio','Julio','Agosto','Septiembre','Octubre','Noviembre','Diciembre']
Autos = ['Suv', 'Sedan', 'Hatchback']

# Datos de ventas específicos para cada producto en cada ciudad
Suv = [120,130,135,140,150,160,170,180,175,185,190,200]
Sedan = [100,90,95,110,120,130,140,145,150,160,170,180]
Hatchback = [80,85,90,95,100,105,110,115,120,125,130,135]

# Configurar el gráfico de barras
fig, ax = plt.subplots(figsize=(15, 6))
bar_width = 0.2
index = np.arange(len(Meses))

# Crear barras para cada producto
plt.bar(index, Suv, bar_width, label='Suv', edgecolor='black')
plt.bar(index + bar_width, Sedan, bar_width, label='Sedan', edgecolor='black')
plt.bar(index + 2 * bar_width,Hatchback , bar_width, label='Hatchback', edgecolor='black')

# Añadir títulos y etiquetas
ax.set_xlabel('Meses')
ax.set_ylabel('Ventas')
ax.set_title('Ventas por automovil por mes')
ax.set_xticks(index + bar_width)
ax.set_xticklabels(Meses)
plt.legend(title='Autos')

# Mostrar el gráfico
plt.show()
```
